Beräkna snölastens karakteristiska värde för byggnaden i figuren samt vertikalreaktionen i vägg A av enbart karakteristisk snölast. Byggnaden är belägen i Kristianstad i område med normal topografi.

## Givet:

```python
s_k = Q(1.5, 'kN/m**2')
#s_k = 1500 * si.N / (si. m**2)
C_e = 1
C_t = 1
mu_2 = 0.4
mu_5 = 0.45
#d = 8 * si.m
d = Q (8, 'm')
#alpha = 45 * deg  #this is from physunits
alpha = Q(45,'deg')
```

```python
%%render
#parameter
s_k
C_e
C_t
mu_2
mu_5
d
alpha
```

# Sökt

Beräkna snölastens karakteristiska värde för byggnaden
samt vertikalreaktionen i vägg A av enbart karakteristisk snölast.

# Lösning

(Obs att snölasten anges per horisontell ytenhet)
(Anta att vindriktning är från höger till vänster)

```python
%%render
s_h = s_k * C_e * C_t * mu_2
s_v = s_k * C_e * C_t * mu_5
```

Moment kring punkt B ger reaktionskraft vid punkt A:

```python
%%render

v_A = (s_v * d / 2 *(d/2+d/4)+s_h*d/2*d/4)/d
```

Obs att den större lasten sv här har placerats på vänstra takhalven eftersom detta är ogynnsammast för lasteffekten på vägg A.
